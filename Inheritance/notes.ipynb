# Inheritance

```python
# Superclass / parent class
class Virus:
    pass

# child class, subclass
class RNAVirus(Virus):
    pass
```

```python
class Virus:
    pass

class RNAVirus(Virus):
    pass

class CoronaVirus(RNAVirus):
    pass

class SARSCov2(CoronaVirus):
    pass
```

```python
RNAVirus.__bases__

# pecial Python attribute that shows the parent class(es) a class inherits from.
```

```python
RNAVirus.__mro__

# shows the order Python follows when looking for methods and attributes in a class hierarchy.
```

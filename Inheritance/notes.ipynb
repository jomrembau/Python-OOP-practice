# Parent Delegation - super()

```python
# super() is used to call methods from a parent class.

# The child class overrides the parent's method, but can still use the parent's version with super() and then add extra behavior.
```

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        super().speak()
        print("Dog barks")

dog = Dog()
dog.speak()
```

# Subclass __init__

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)

dog = Dog("Buddy")
print(dog.name)
```

# Subclassing properties. 


```python
# Short explanation:

# Animal.name = the property from the parent class.
# .setter = replace its setter method.
# The subclass creates a new setter for the inherited property.
```

```python
class Animal:
    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        print("Animal setter called")
        self._name = value


class Dog(Animal):
    @Animal.name.setter
    def name(self, value):
        print("Dog setter called")
        self._name = value.upper()


dog = Dog()
dog.name = "buddy"

print(dog.name)
```

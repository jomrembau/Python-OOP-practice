```python
class Book:
    def __init__(self, title, author, book_type, pages):
        self.title = title
        self.author = author
        self.book_type = book_type
        self.pages = pages

    def __repr__(self):
        return f'Book("{self.title}", "{self.author}", "{self.book_type}", {self.pages})'

    def __str__(self):
        return f"{self.title} by {self.author} in {self.book_type}"
```

```python
book1 = Book("Art of War", "Sun Tzu", "Hardcover", "120")
```

```python
str(book1)
```

```python
class Book:
    def __init__(self, title, author, book_type, pages):
        self.title = title
        self.author = author
        self.book_type = book_type
        self.pages = pages

    def __repr__(self):
        return f'Book("{self.title}", "{self.author}", "{self.book_type}", {self.pages})'

    def __format__(self, format_spec):
        if format_spec == "short":
            return f"{self.title}"
```

```python
book1 = Book("Art of War", "Sun Tzu", "Hardcover", "120")
book2 = Book("Art of War", "Sun Tzu", "Hardcover", "120")
```

```python
id(book1)
```

```python
id(book2)
```

```python
class Book:
    def __init__(self, title, author, book_type, pages):
        self.title = title
        self.author = author
        self.book_type = book_type
        self.pages = pages

    def __repr__(self):
        return f'Book("{self.title}", "{self.author}", "{self.book_type}", {self.pages})'

    def __eq__(self,other):
        return True        
```

```python
book1 = Book("Art of War", "Sun Tzu", "Hardcover", "120")
book2 = Book("Art of War", "Sun Tzu", "Hardcover", "120")
```

```python
book1 == book2
```

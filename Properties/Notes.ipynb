```python
class Customer:
    
    def __init__(self, loyalty):
        self.loyalty = loyalty
```

```python
c1 = Customer("bronze")
c2 = Customer("gold")
c3 = Customer("platinum")
```

```python
def get_discount(customer):
    discounts = {
        "bronze": .1,
        "gold": .2,
        "platinum": .35
    }

    discount = discounts.get(customer.loyalty, None)

    if not discount:
        raise ValueError("Could not determine the customer's discount")

    return discount

```

```python
for customer in(c1,c2,c3):
    print(f"{get_discount(customer):.0%}")
```

# Decorator

What does a Decorator do?

A decorator adds extra behavior to a function without changing the original function code.

Think of it like a wrapper around a gift:

Original function = gift.
Decorator = wrapping paper that adds something extra.

```python
# The getter is decorated with @property, and the setter is decorated with @property_name.setter
```

```python
def greet(who):
    how = "Good Morning, "

    def create_greeting():
        print(f"{how}{who}")

    return create_greeting


        
```

```python
a = greet("jomir")
```

```python
a()
```

```python
import random

def even_or_odd(func):
    def inner():
        num = func()
        print(f"The number is {'even' if num % 2 == 0 else 'odd'}")
        return num
    return inner

@even_or_odd
def bingo():
    return random.randint(1, 100)

bingo()
```

```python
@even_or_odd
def bingo():
    return randint(1,47)
```

```python
# property(fget,fset,fdel)
```

```python
class Customer:
    loyalty_levels = {"bronze", "gold", "platinum"}

    def __init__(self, loyalty):
        self._loyalty = loyalty

    def loyalty_getter(self):
        return self._loyalty

    def loyalty_setter(self, level):
        if level in self.__class__.loyalty_levels:
            self._loyalty = level
        else:
            raise ValueError(f"Invalid loyalty {level}")

    def loyalty_deleter(self):
        del self._loyalty

    loyalty = property(fget=loyalty_getter, fset=loyalty_setter, fdel=loyalty_deleter)
```

# same as

```python
class Customer:
    loyalty_levels = {"bronze", "gold", "platinum"}

    def __init__(self, loyalty):
        self._loyalty = loyalty

    @property
    def loyalty(self):
        return self._loyalty

    @loyalty.setter
    def loyalty(self, level):
        if level in self.__class__.loyalty_levels:
            self._loyalty = level
        else:
            raise ValueError(f"Invalid loyalty {level}")

    @loyalty.deleter
    def loyalty(self):
        del self._loyalty
```

```python
class MercedezBenz:
    doors = 4
    wheel = 4
    model = "G"

    def __init__(self, color):
        self.color = color
    
    def drive(self):
        return f"A Mercedes is driving. And it is {self}\n"
```

```python
m1 = MercedezBenz("black")
m2 = MercedezBenz("red")
```

```python
getattr(m1, "color")
```

```python
setattr(m1, "color", "yellow")
```

```python
getattr(m1, "color")
```

```python
objs = [m1,m2]

attribs = ["color", "doors"]
values = ["blue",3]

```

```python
for obj in objs:
    for attrib, val in zip(attribs,values):
        setattr(obj, attrib, val)
```

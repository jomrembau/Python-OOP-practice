


CLASS AND STATIC METHODS



```python
class MercedezBenz:
    doors = 4
    wheel = 4
    model = "G"

    def __init__(self, color):
        self.color = color
    
    def drive(self):
        return f"A Mercedes is driving. And it is {self}\n"

    @staticmethod
    def auto_drive():
        return "Auto-Driving"

    @classmethod
    def create_lease(cls):
        print(f"A lease for {clas} will be created.")

    
```

```python
#Alternative Syntax

class MercedezBenz:
    doors = 4
    wheel = 4
    model = "G"

    def __init__(self, color):
        self.color = color
    
    def drive(self):
        return f"A Mercedes is driving. And it is {self}\n"

    def auto_drive():
        return "Auto-Driving"

    def create_lease(cls):
        print(f"A lease for {clas} will be created.")

    create_lease = classmethod(create_lease)
    auto_drive = staticmethod(auto_drive)

    
```
